```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import math as mt
import sympy as sp 
def sen(grados):
    return mt.sin(mt.radians(grados))
def cos(grados):
    return mt.cos(mt.radians(grados))
def acos(grados):
    return (mt.acos(grados))*(180/mt.pi)
def asen(grados):
    return (mt.asin(grados))*(180/mt.pi)
l=4
b=2.5
a=asen(b/l)
c=180-a
Phi_punto=1*np.pi/180
```

```python
Phi=np.linspace(0, 360, 1000).reshape(-1, 1)
Theta=[]
Theta_punto=[]
for i in Phi:
    R=acos((l*cos(i))/(np.sqrt((-b+l*sen(i))**2+(l*cos(i))**2)))
    DR=(((-l*sen(i))/(np.sqrt((-b+l*sen(i))**2+(l*cos(i))**2)))-((l*cos(i)*(2*l*cos(i)*(l*sen(i)-b)-2*l*l*sen(i)*cos(i)))/(2*(((l*sen(i)-b)**2)+(l*cos(i))**2)**(3/2))))/((np.sqrt(1-(l*cos(i))**2/((l*sen(i)-b)**2+(l*cos(i))**2))))
    if (i<a):
        Theta.append(-R)
        Theta_punto.append(DR*Phi_punto)
    if (i>=a and i<c):
        Theta.append(R)
        Theta_punto.append(-DR*Phi_punto)
    if (i>=c):
        Theta.append(360-R)
        Theta_punto.append(DR*Phi_punto)
```

```python
plt.plot(Phi,Theta)
plt.xticks(range(0,360,40))
plt.yticks((range(-30,340,40)))
plt.title('Phi Vs Theta')
plt.ylabel('Grados (°)', fontsize=12)
plt.xlabel('Grados (°)', fontsize=12)
plt.show
```

```python
plt.plot(Phi,Theta_punto)
plt.title('Phi Vs Theta_punto')
plt.ylabel('Rad/s (°)', fontsize=12)
plt.xlabel('Grados (°)', fontsize=12)
plt.show
```
